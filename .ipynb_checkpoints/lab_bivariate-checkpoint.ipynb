```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
df
```

```python
df = df[df.columns[df.isnull().mean() < 0.8]]
```

```python
df.drop("uid", inplace=True, axis=1)
```

```python
potential_categorical_from_numerical = df.select_dtypes("number").loc[:, df.select_dtypes("number").nunique() < 20]
potential_categorical_from_numerical
```

```python
df_categorical = pd.concat([df.select_dtypes("object"), potential_categorical_from_numerical], axis=1)

df_numerical = df.select_dtypes("number").drop(columns=potential_categorical_from_numerical.columns)
```

```python
crosstab_result = pd.crosstab(df['category'], df['isBestSeller']).sort_values(by=True, ascending=False).head(10)
crosstab_result
```

```python
#Groceries and Health & Personal Care have a lot of best-sellers, which makes sense because these are necessities in life
```

```python
from scipy.stats import chi2_contingency

chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
```

```python
# p-value is 0, indicating strong evidence against the null hypothesis
```

```python
from scipy.stats.contingency import association

association(crosstab_result, method="cramer")
```

```python
# The value of 0.17 suggests a weak association between 'isBestSeller' and 'category'
```

```python
crosstab_result.plot(kind="bar", stacked=True)
```

```python
# This chart shows the dominance in Sports & Outdoors products
```

```python
Q3 = df['price'].quantile(0.75)
Q1 = df['price'].quantile(0.25)
IQR = Q3 - Q1
lower_wisker = Q1 - 1.5 * IQR
upper_wisker = Q3 + 1.5 * IQR

df = df[(df['price'] >= lower_wisker) & (df['price'] <= upper_wisker)]
df
```

```python
top_20 = df['category'].value_counts().head(20).index
```

```python
top_20_df = df[df['category'].isin(top_20)]
```

```python
plt.figure(figsize=(18, 16))
sns.violinplot(data=top_20_df, x='price', y='category', palette="coolwarm")
```

```python
# Category 'Men' has the highest median, based on the white dot
```

```python
top_10 = df['category'].value_counts().head(10).index
```

```python
top_10_df = df[df['category'].isin(top_10)]
```

```python
price_mean = top_10_df.groupby('category')['price'].mean().head(10).reset_index()
```

```python
price_mean = price_mean.sort_values('price', ascending=False)
```

```python
sns.barplot(data=price_mean, x='price', y='category', palette="coolwarm")
```

```python
# Sports & Outdoors has the highest average price
```

```python
sns.boxplot(data=top_10_df, x='price', y='category', palette="coolwarm")
```

```python
# Sports & Outdoors has the highest median price
```
